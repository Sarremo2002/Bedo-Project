```python
import pandas as pd




df = pd.read_csv('Most Streamed Spotify Songs 2024.csv', encoding='ISO-8859-1')


df.head()


```

```python
df.info()
```

```python
print("Missing values per column:\n", df.isnull().sum())
```

```python
empty_cols = df.columns[df.isnull().all()]
print("Fully empty columns:\n", empty_cols)
```

```python
df = df.drop(columns=empty_cols)
```

```python
# show the updated table
df.head()
```

```python
# Show the number missing values of each row for its respective column
df.isnull().sum()

```

```python
# shows the rows containing the missing values
df[df.isnull().any(axis=1)]
```

```python
#fill in missing values for columns of catogorial data and numerical data

cat_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(include='number').columns


for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)
        print(f"Filled '{col}' with mode: {mode_val}")


for col in num_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Filled '{col}' with median: {median_val}")
```

```python
# proof there are no missing values left
df.isnull().sum()
```

```python
# gives us the summary of numerical data like mean values and number of non empty fields
df.describe()
```

```python
# make all column names lower case and add and underscore between spaces
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['spotify_streams'], bins=30, kde=True)
plt.title("Distribution of Spotify Streams")
plt.xlabel("Spotify Streams")
plt.ylabel("Count")
plt.show()
```

```python
# Remove commas and convert to numeric
df['spotify_streams'] = df['spotify_streams'].str.replace(',', '')
df['spotify_streams'] = pd.to_numeric(df['spotify_streams'], errors='coerce')
```

```python
# Remove top 1%
q99 = df['spotify_streams'].quantile(0.99)
df_filtered = df[df['spotify_streams'] <= q99]

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_filtered['spotify_streams'], bins=30, kde=True)
plt.title("Spotify Streams Distribution (Without Top 1%)")
plt.xlabel("Spotify Streams")
plt.ylabel("Count")
plt.show()
```

```python
# List of columns that should be numeric but may have commas
columns_to_clean = [
    'spotify_streams', 'spotify_playlist_reach', 'youtube views',
    'youtube likes', 'tiktok posts', 'tiktok likes', 'tiktok views',
    'youtube playlist reach', 'pandora streams', 'pandora track stations',
    'soundcloud streams', 'shazam counts'
]

# Remove commas and convert to numeric
for col in columns_to_clean:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace(',', '').str.strip()
        df[col] = pd.to_numeric(df[col], errors='coerce')

```

```python
# Re-identify numeric columns (only columns that exist and are numeric)
num_cols = df.select_dtypes(include='number').columns


print("✅ All numeric columns have been normalized between 0 and 1.")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Fit and transform only the existing numeric columns
df[num_cols] = scaler.fit_transform(df[num_cols])

print("✅ All numeric columns have been normalized between 0 and 1.")
```

```python
df[num_cols].describe()
```

```python

q99 = df['spotify_streams'].quantile(0.99)
df_filtered = df[df['spotify_streams'] <= q99]

import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_filtered['spotify_streams'], bins=30, kde=True)
plt.title("Spotify Streams Distribution (Without Top 1%)")
plt.xlabel("Spotify Streams")
plt.ylabel("Count")
plt.show()
```

```python
# List of columns to drop
columns_to_drop = ['album_name', 'release_date', 'isrc', 'artist']

# Drop them if they exist
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)

print("✅ Removed: album name, release date, isrc, artist")
```

```python
# Save the cleaned and normalized DataFrame
df.to_csv("final_spotify_data.csv", index=False)

# Download it
from google.colab import files
files.download("final_spotify_data.csv")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load your dataset
df = pd.read_csv('final_spotify_data-2.csv')

# List of columns that are strings but represent large numeric values
columns_to_normalize = [
    'spotify_playlist_count',
    'youtube_views',
    'youtube_likes',
    'tiktok_posts',
    'tiktok_likes',
    'tiktok_views',
    'youtube_playlist_reach',
    'airplay_spins',
    'siriusxm_spins',
    'deezer_playlist_reach',
    'pandora_streams',
    'pandora_track_stations',
    'soundcloud_streams',
    'shazam_counts'
]

# Step 1: Remove commas and convert to float
for col in columns_to_normalize:
    df[col] = df[col].str.replace(',', '', regex=False).astype(float)

# Step 2: Normalize using Min-Max scaling
scaler = MinMaxScaler()
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

# Optional: Save the cleaned and normalized data
df.to_csv('spotify_data_normalized.csv', index=False)

# Display the first few rows to confirm
df.head()
```
